```python
#Implementing the youtube api
api_service_name = "youtube"
api_version = "v3"

# Get credentials and create an API client
youtube = build(
    api_service_name, api_version, developerKey=api_key)
```

```python
#Function to iterate through all the ids on the list and extract the main info
import math

def get_channel_intro_stats(youtube, intro_ids):
    
    all_data = []
    #creating chunks
    chunks_intro = [intro_ids[x:x+50] for x in range(0, len(intro_ids), 50)]
    
 
    for chunk_intro in chunks_intro:
        request = youtube.channels().list(
            part="snippet,contentDetails,statistics",
            id=','.join(chunk_intro)
        )
        response = request.execute()

        #looping through the chunks
        for item in response['items']:
            data = {'channelName': item['snippet']['title'],
                    'subscribers': item['statistics']['subscriberCount'],
                    'views': item['statistics']['viewCount'],
                    'totalViews': item['statistics']['videoCount'],
                    'playlistID': item['contentDetails']['relatedPlaylists']['uploads']

            }

            all_data.append(data)

    return pd.DataFrame(all_data)
```

```python
#Code to extract video id's from one sole youtube channel id

playlist_id = 'UUX6OQ3DkcsbYNE6H8uQQuVA'

def get_video_intro_ids(youtube, playlist_id):
    
    video_intro_ids = []
    
    request = youtube.playlistItems().list(
        part="snippet,contentDetails",
        playlistId = playlist_id,
        maxResults = 50
    )
    response = request.execute()

    for item in response['items']:
        video_intro_ids.append(item['contentDetails']['videoId'])
        
    next_page_token = response.get('nextPageToken')
    while next_page_token is not None:
        request = youtube.playlistItems().list(
                    part='contentDetails',
                    playlistId = playlist_id,
                    maxResults = 50,
                    pageToken = next_page_token)
        response = request.execute()

        for item in response['items']:
            video_intro_ids.append(item['contentDetails']['videoId'])
            
        next_page_token = response.get('nextPageToken')
    
    return video_intro_ids
```

```python
video_intro_ids = get_video_intro_ids(youtube, playlist_id)
```

```python
# to retrieve transcription of the first 15 seconds of videos, for videos posted starting on 2018/01/01
# giving exception to videos that can't be transcribed, returning as "no transcription available"

import datetime
import googleapiclient.discovery
from youtube_transcript_api import YouTubeTranscriptApi
#function to return only videos posted after a specific datetime

def get_video_publication_date(video_id, youtube):
    request = youtube.videos().list(
        part="snippet",
        id=video_id
    )
    response = request.execute()
    if "items" in response and len(response["items"]) > 0:
        snippet = response["items"][0]["snippet"]
        published_at = snippet["publishedAt"]
        return datetime.datetime.strptime(published_at, "%Y-%m-%dT%H:%M:%SZ")
    return None

# function to get the transcription

def get_first_15_seconds_transcription(videos_list, api_key):
    youtube = googleapiclient.discovery.build("youtube", "v3", developerKey=api_key)
    transcripts = []
    for video_id in videos_list:
        publication_date = get_video_publication_date(video_id, youtube)
        if publication_date is None or publication_date < datetime.datetime(2018, 1, 1):
            continue
        try:
            transcript = YouTubeTranscriptApi.get_transcript(video_id)
        except:
            transcript = None
        if transcript is not None:
            for i, part in enumerate(transcript):
                if part["start"] >= 15.0:
                    transcripts.append((" ".join([p["text"] for p in transcript[:i]]), video_id))
                    break
            else:
                transcripts.append((" ".join([p["text"] for p in transcript]), video_id))
        else:
            transcripts.append(("no transcription available", video_id))
    return transcripts

# call the transcription function

transcriptions = get_first_15_seconds_transcription(video_intro_ids, api_key)
transcription_df = pd.DataFrame(transcriptions, columns=["Transcription", "Video ID"])
transcription_df
```
