```python
#Implementing the youtube api
api_service_name = "youtube"
api_version = "v3"

# Get credentials and create an API client
youtube = build(
    api_service_name, api_version, developerKey=api_key)
```

```python
#Function to iterate through all the ids on the list and extract the main info
import math

def get_channel_stats(youtube, channel_ids):
    
    all_data = []
    #creating chunks
    chunks = [channel_ids[x:x+50] for x in range(0, len(channel_ids), 50)]
    
 
    for chunk in chunks:
        request = youtube.channels().list(
            part="snippet,contentDetails,statistics",
            id=','.join(chunk)
        )
        response = request.execute()

        #looping through the chunks
        for item in response['items']:
            data = {'channelName': item['snippet']['title'],
                    'subscribers': item['statistics']['subscriberCount'],
                    'views': item['statistics']['viewCount'],
                    'totalViews': item['statistics']['videoCount'],
                    'playlistID': item['contentDetails']['relatedPlaylists']['uploads']

            }

            all_data.append(data)

    return pd.DataFrame(all_data)
```

```python
#show main channel stats in a dataframe
channel_stats = get_channel_stats(youtube, channel_ids)
channel_stats
```

```python
#Converting the playlist into list values so they can be iterated on
playlistID_list = channel_stats.playlistID.values.tolist()
print(playlistID_list)
```

```python
#Function that gets a lisft of video ids from the list of youtube channel ids
def get_video_ids(youtube, playlistID_list):

    video_ids = []

    for playlist_id in playlistID_list:
        request = youtube.playlistItems().list(
            part="snippet,contentDetails",
            playlistId = playlist_id,
            maxResults = 50
        )
        response = request.execute()

        for item in response['items']:
            video_ids.append(item['contentDetails']['videoId'])

        next_page_token = response.get('nextPageToken')
        while next_page_token is not None:
            request = youtube.playlistItems().list(
                        part='contentDetails',
                        playlistId = playlist_id,
                        maxResults = 50,
                        pageToken = next_page_token)
            response = request.execute()

            for item in response['items']:
                video_ids.append(item['contentDetails']['videoId'])

            next_page_token = response.get('nextPageToken')

    return video_ids
video_ids = get_video_ids(youtube, playlistID_list)
```

```python
#Function to get the remaining details for each youtube video, iterating through the entire list
def get_video_details(youtube, video_ids):

    all_video_info = []
    
    for i in range(0, len(video_ids), 50):
        request = youtube.videos().list(
            part="snippet,contentDetails,statistics",
            id=','.join(video_ids[i:i+50])
        )
        response = request.execute() 

        for video in response['items']:
            stats_to_keep = {'snippet': ['channelTitle', 'title', 'tags', 'publishedAt', 'categoryId'],
                             'statistics': ['viewCount', 'likeCount', 'commentCount'],
                             'contentDetails': ['duration', 'definition']
                            }
            video_info = {}
            video_info['video_id'] = video['id']

            for k in stats_to_keep.keys():
                for v in stats_to_keep[k]:
                    try:
                        video_info[v] = video[k][v]
                    except:
                        video_info[v] = None

            all_video_info.append(video_info)
    
    return pd.DataFrame(all_video_info)
video_df = get_video_details(youtube, video_ids)
```
