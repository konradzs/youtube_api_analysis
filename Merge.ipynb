```python
#Polarity scores to define a sentiment analysis
def polarity_scores_roberta(example):
    encoded_text = tokenizer(example, return_tensors='pt')
    output = model(**encoded_text)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    scores_dict = {
        'roberta_neg' : scores[0],
        'roberta_neu' : scores[1],
        'roberta_pos' : scores[2]
    }
    return scores_dict
```

```python
#Function to count the amount of times a sentence started with a specific string

def count_first_sentence_matches(column):
    
    # Initialize dictionaries to store counts
    string_counts = {}
    pattern_counts = {
        'This is': 0,
        'I am': 0,
        'We are': 0,
        'We + verb': 0,
        'I + verb': 0,
        'We + word + verb': 0,
        'I + word + verb': 0,
       
    }
    
    begin_words = []  # Initialize list to store matched sentences
    
    for string in column:
        doc = nlp(string)
        
        # Get the first sentence of the document
        first_sent = next(doc.sents)
        
        # Check if the first sentence matches one of the specified patterns
        first_word = first_sent[0].text.lower()
        second_word = first_sent[1].text.lower()
        third_word = None if len(first_sent) < 3 else first_sent[2].text.lower()
        
        if (first_word == 'this' and second_word == 'is'):
            pattern_counts['This is'] += 1
            begin_words.append('This is')
            string_counts[string] = 1
        elif (first_word == 'i' and second_word == 'am'):
            pattern_counts['I am'] += 1
            begin_words.append('I am')
            string_counts[string] = 1
        elif (first_word == 'we' and second_word == 'are'):
            pattern_counts['We are'] += 1
            begin_words.append('We are')
            string_counts[string] = 1
        elif (first_word == 'we' and first_sent[1].pos_ == 'VERB'):
            pattern_counts['We + verb'] += 1
            begin_words.append('We + verb')
            string_counts[string] = 1
        elif (first_word == 'i' and first_sent[1].pos_ == 'VERB'):
            pattern_counts['I + verb'] += 1
            begin_words.append('I + verb')
            string_counts[string] = 1
        elif (first_word == 'we' and first_sent[2].pos_ == 'VERB'):
            pattern_counts['We + word + verb'] += 1
            begin_words.append('We + word + verb')
            string_counts[string] = 1
        elif (first_word == 'i' and first_sent[2].pos_ == 'VERB'):
            pattern_counts['I + word + verb'] += 1
            begin_words.append('I + word + verb')
            string_counts[string] = 1
        else:
            begin_words.append('')
            string_counts[string] = 0
    
    # Add begin_words as a new column to the dataframe
    beast_df['begin_words'] = begin_words
    
    return string_counts, pattern_counts
```

```python
transcription_column = beast_df['transcription']

# Call the count_first_sentence_matches function on the transcription column
counts = count_first_sentence_matches(transcription_column)
```
